```python

"""
Methods:
get_name()
get_courses()
add_course(course)
drop_course(course)
shares_course_with(other)
get_number_of_students() (must use @classmethod)
Rules:

number_of_students increases every time a Student object is created.
get_number_of_students() returns the number_of_students
add_course() must place the course inside  _courses
drop_course() must remove the course from inside  _courses
shares_course_with() requirement : you must use a set operation
"""

class Student:
  number_of_students = 0

  def __init__(self, name):
    self._name = name
    self._courses = set()
    Student.number_of_students += 1

  def get_name(self):
    return self._name

  def get_courses(self):
    return self._courses

  def add_course(self, course):
    self._courses.add(course)

  def drop_course(self, course):
    self._courses.remove(course)

  def shares_course_with(self, other):
    for item in self._courses:
      if item in other._courses:
        return True
    return False

  @classmethod
  def get_number_of_students(cls):
    return cls.number_of_students

s1 = Student("Avi")
s2 = Student("Leah")


s1.add_course("Math")
s1.add_course("CS")
s1.add_course("Math")
s2.add_course("Bio")
s2.add_course("Math")

print(s1.get_name())
print(s1.get_courses())
print(s1.shares_course_with(s2))
print(Student.get_number_of_students())
```
