```python
class Student:
  number_of_students = 0

  def __init__(self, name):
   self._name = name
   self._courses = set()
   Student.number_of_students += 1

  def get_name(self):
    return self._name


  def get_courses(self):
    return self._courses

  def add_course(self, course):
    self._courses.add(course)


  def drop_course(self, course):
    self._courses.remove(course)


  def shares_course_with(self, other):
    for item in self._courses:
      if item in other._courses:
        return True
    return False


  @classmethod
  def get_number_of_students(cls):
    return cls.number_of_students

s1 = Student("Avi")
s2 = Student("Leah")


s1.add_course("Math")
s1.add_course("CS")
s1.add_course("Math")
s2.add_course("Bio")
s2.add_course("Math")

print(s1.get_name())
print(s1.get_courses())
print(s1.shares_course_with(s2))
print(Student.get_number_of_students())
```

```python
"""
Return a dictionary mapping each student’s name to how many other students they share at least one course with.
"""

def connection_counts(students):
  connections = {}

  for student in students:
    shared_classmates = set()
    connections[student.get_name()] = 0
    for potential_classmate in students:
      if potential_classmate != student:
        if student.shares_course_with(potential_classmate):
          if potential_classmate not in shared_classmates:
            shared_classmates.add(potential_classmate.get_name())
    for classmate in shared_classmates:
      connections[student.get_name()] += 1

  return connections

s1 = Student("Avi")
s2 = Student("Leah")
s3 = Student("Noah")
s4 = Student("Sara")

s1.add_course("Math")
s1.add_course("CS")
s2.add_course("Math")
s2.add_course("Bio")
s3.add_course("CS")
s3.add_course("History")
s4.add_course("Art")

students = [s1, s2, s3, s4]

print(connection_counts(students))

```

```python
def build_course_roster(students):
  roster = {}

  for student in students:
    for cls in student.get_courses():
      if cls not in roster:
        roster[cls] = set()
      roster[cls].add(student.get_name())
  return roster

s1 = Student("Avi")
s2 = Student("Leah")
s3 = Student("Noah")

s1.add_course("Math")
s1.add_course("CS")
s2.add_course("Math")
s2.add_course("Bio")
s3.add_course("CS")
s3.add_course("History")

students = [s1, s2, s3]

print(build_course_roster(students))
```
