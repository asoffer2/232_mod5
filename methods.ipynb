```python
class Animal:
  def __init__(self, name):
    self.name = name

  def speak(self):
    print("HELLO")

class Dog(Animal):
  def __init__(self, age, name):
    super().__init__(name)
    self.age = age

  def speak(self):
    print("BOW WOW")
    print(super().speak())

  @staticmethod
  def is_age_valid(age: int):
    return 0 < age < 18

  @staticmethod
  def is_female(gender):
    return gender == "Female"

my_animal = Dog(12, "Bonzo")
print(my_animal.age, my_animal.name)
my_animal.speak()
print(Dog.is_age_valid(15))
```

```python
from abc import ABC, abstractmethod, abstractclassmethod

class Animal(ABC): # ABC means it will have an abstract method
  animal_counts = 0
  def __init__(self):
    print("I am forcing you to implement these methods")
    Animal.animal_counts += 1

  @abstractmethod
  def run(self):
    pass

  @abstractmethod
  def speak(self):
    pass

  @abstractclassmethod
  def animal_count(cls):
    pass

class Dog(Animal):

  def run(self):
    pass


  def speak(self):
    pass

  def animal_count(cls):
    return Dog.animal_counts

my_dog = Dog()
my_dog.speak()
print(my_dog.animal_count())


```

```python
from abc import ABC, abstractmethod

class WordGameController(ABC):

  @abstractmethod
  def play(self):
    pass

  @abstractmethod
  def make_guess(self, letter):
    pass

  @abstractmethod
  def check_game_over(self):
    pass

```

```python
class TinyWordGame(WordGameController):
  def __init__(self, secret_word):
    self.secret_word = secret_word.lower()
    self.guesses = set()

  def play(self):
    print("Welcome to the game, try to guess one letter at a time")

  def make_guess(self, letter):
    letter = letter.lower()
    self.guesses.add(letter)
    return letter in self.secret_word

  def check_game_over(self):
    return all(letter in self.guesses for letter in self.secret_word)

game = TinyWordGame("cat")
game.play()
print(game.make_guess("c"))
print(game.check_game_over())
```

```python
class HintWordGame(WordGameController):
  def __init__(self, secret_word, hint):
    self.secret_word = secret_word
    self.hint = hint
    self.guesses = set()

  def play(self):
    print("Welsome to the game, here is a hint", self.hint)


  def make_guess(self, letter):
    letter = letter.lower()
    self.guesses.add(letter)
    return letter in self.secret_word


  def check_game_over(self):
    return all(letter in self.guesses for letter in self.secret_word)

game = HintWordGame("dog", "A floppy companion")
game.play()
print(game.make_guess("c"))
print(game.check_game_over())
```

```python
class LimitWordGame(WordGameController):
  max_attempts = 5
  def __init__(self, secret_word):
    self.secret_word = secret_word
    self.attempts = 0
    self.guesses = set()


  @classmethod
  def play(cls):
    print(f"Welcome to limit word game, you have {cls.max_attempts} attemps")

  def make_guess(self, letter):
    letter = letter.lower()
    self.guesses.add(letter)
    self.attempts += 1
    return letter in self.secret_word

  def check_game_over(self):
    return all(letter in self.guesses for letter in self.secret_word or self.attempts > self.max_attempts)

game = LimitWordGame("hello")
game.play()
print(game.make_guess("h"))
print(game.check_game_over())



```

```python
class Shape:
  shape_count = 0
  def __init__(self, color):
    self.color = color
    self.shape_count += 1

  @classmethod
  def get_shape_count(cls):
    return cls.shape_count

  @staticmethod
  def is_valid_color(color):
    return isinstance(color, str)

  @abstractmethod
  def area(self):
    pass

  @abstractmethod
  def describe(self):
    pass

```

```python
class Circle(Shape):
  def __init__(self, color, radius):
    super().__init__(color)
    self.radius = radius

  def area(self):
    return self.radius*2

  def describe(self):
    return f"I am a {self.color} circle"

```

```python
class Square(Shape):
  def __init__(self, color, width):
    super().__init__(color)
    self.width = width

  def area(self):
    return self.width*2

  def describe(self):
    return f"I am a {self.color} square"

circle = Circle("blue", 5)
square = Square("yellow", 7)

print(circle.area(), circle.describe())
print(square.area(), square.describe())
print(circle.get_shape_count())

```
