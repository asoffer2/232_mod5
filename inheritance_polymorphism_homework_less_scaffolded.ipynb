## Problem 1: Build an Animal Hierarchy

Create a parent class named `Animal`.

### The `Animal` class must:
- store a name
- store an age
- include a method named `speak()` that returns a general animal sound
- include a method named `move()` that returns a general description of movement
- include a method named `describe()` that returns a short sentence about the animal

Then create these child classes:

- `Dog`
- `Cat`
- `Rabbit`

### Child class requirements
Each child class must:
- inherit from `Animal`
- use the parent constructor appropriately
- override `speak()`
- override `move()`

### After creating your classes
Create one object of each child class and demonstrate that:
- each object stores its own data
- each object gives a different result for `speak()`
- each object gives a different result for `move()`
- `describe()` works for each object


```python
# Write your code for Problem 1 here.
class Animal:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def speak(self):
    return "Hello"

  def move(self):
    return "Moves forward"

  def describe(self):
    return f"This animal, {self.name}, has fur, is {self.age} years old and is a kind of pet."

class Dog(Animal):
  def __init__(self, name, age):
    super().__init__(name, age)

  def speak(self):
    return "Woof Woof"

  def move(self):
    return "Runs"

  def __str__(self):
    return "Dog"

class Cat(Animal):
  def __init__(self, name, age):
    super().__init__(name, age)

  def speak(self):
    return "Meow meow"

  def move(self):
    return "Slinks forward"

  def __str__(self):
    return "Cat"

class Rabbit(Animal):
  def __init__(self, name, age):
    super().__init__(name, age)

  def speak(self):
    return "Maaaaa"

  def move(self):
    return "Hops"

  def __str__(self):
    return "Rabbit"

my_cat = Cat("Fluffy", 3)
my_dog = Dog("Oreo", 6)
my_rabbit = Rabbit("Snowy", 2)

pets = [my_cat, my_dog, my_rabbit]

for pet in pets:
  print(f"{pet}:")
  print(f"{pet.speak()}, {pet.move()}, {pet.describe()}")
  print(f"{pet.name}, {pet.age}")




```

## Problem 2: Build a Vehicle Hierarchy

Create a parent class named `Vehicle`.

### The `Vehicle` class must:
- store a brand
- store a speed
- include a method named `move()` that returns a general statement about transportation
- include a method named `sound_horn()` that returns a general horn sound
- include a method named `describe()` that returns a short sentence about the vehicle

Then create these child classes:

- `Car`
- `Truck`
- `Motorcycle`

### Child class requirements
Each child class must:
- inherit from `Vehicle`
- use the parent constructor appropriately
- override `move()`
- override `sound_horn()`

### After creating your classes
Create one object of each child class and demonstrate that:
- each object stores its own data
- each object gives a different result for `move()`
- each object gives a different result for `sound_horn()`
- `describe()` works for each object


```python
# Write your code for Problem 2 here.
class Vehicle:
  def __init__(self, brand, speed):
    self.brand = brand
    self.speed = speed

  def move(self):
    return "Gets you from one place to the next."

  def sound_horn(self):
    return "honk honk"

  def describe(self):
    return f"This {self.brand} vehicle has an engine and a speed of {self.speed}."

class Car(Vehicle):
  def __init__(self, brand, speed):
    super().__init__(brand, speed)

  def move(self):
    return "Moves on four wheels and not too big."

  def sound_horn(self):
    return "beeeeeeep"

  def __str__(self):
    return "Car"

class Truck(Vehicle):
  def __init__(self, brand, speed):
    super().__init__(brand, speed)

  def move(self):
    return "Big and must drive under 70 mph."

  def sound_horn(self):
    return "HOOOOOOOOONK"

  def __str__(self):
    return "Truck"

class Motorcycle(Vehicle):
  def __init__(self, brand, speed):
    super().__init__(brand, speed)

  def move(self):
    return "Two wheels and can weave through traffic"

  def sound_horn(self):
    return "ding ding"

  def __str__(self):
    return "Motorcycle"

my_car = Car("Honda", 60)
my_truck = Truck("Ford", 40)
my_motorcycle = Motorcycle("Harley", 80)

vehicles = [my_car, my_truck, my_motorcycle]

for vehicle in vehicles:
  print(f"{vehicle}: brand - {vehicle.brand}, speed - {vehicle.speed}")
  print(f"{vehicle.move()}, {vehicle.sound_horn()}, {vehicle.describe()}")

```
